```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

# Define the function
def generate_uniform_numbers(bottom, ceiling, count):
    return uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)

# Call the function with the given parameters
data1 = generate_uniform_numbers(10, 15, 100)
data2 = generate_uniform_numbers(10, 60, 1000)

# Plot the histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# First histogram
ax1.hist(data1, bins=10, density=False, alpha=0.7, color="g")
ax1.set_title('Uniform Distribution [10, 15]')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')
ax1.set_yticks(range(0, 110, 10))

# Second histogram
ax2.hist(data2, bins=10, density=False, alpha=0.7, color="b")
ax2.set_title('Uniform Distribution [10, 60]')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')
ax2.set_yticks(range(0, 110, 10))

plt.tight_layout()
plt.show()
```

```python
# numpy

def Numeros_aleatorios(bottom,ceiling,count):
  aleatoriosNp = np.random.uniform(bottom, ceiling, count)
  return aleatoriosNp
ejemplo1 = Numeros_aleatorios(10,15,100)
ejemplo2 = Numeros_aleatorios(10,60,1000)

# Plot the histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# first plot
ax1.hist(ejemplo1, bins=10, density=False, alpha=0.7, color="g")
ax1.set_title('Uniform Distribution [10, 15]')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')
ax1.set_yticks(range(0, 110, 10))

# Second histogram
ax2.hist(ejemplo2, bins=10, density=False, alpha=0.7, color="b")
ax2.set_title('Uniform Distribution [10, 60]')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')
ax2.set_yticks(range(0, 110, 10))

plt.tight_layout()
plt.show()
```

## Distribución Normal

1. De la misma manera en el reto de Distribución Uniforme, escriba una función que genere números aleatorios distribuidos normalmente.
1. Genera 1.000 números distribuidos normalmente con media de 10 y desviación típica de 1
1. Genera 1.000 números distribuidos normalmente con una media de 10 y una desviación típica de 50
2. Representa gráficamente las distribuciones de los datos generados.

Resultado esperado:



```python
from scipy.stats import norm 
```

```python
def normal_distribution(mean, std_dev, size):
    return norm.rvs(loc=mean, scale=std_dev, size=size)
```

```python
data_a = normal_distribution(10, 1, 1000)
data_b = normal_distribution(10, 50, 1000)
```

```python
# plot histogram
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))

# first plot
ax1.hist(data_a, bins=60, density=False, alpha=0.7, color="g", label=f'Mean=10, Std Dev=1')
ax1.set_title('Normal Distribution [Mean=10, Std Dev=1]')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')
ax1.legend()

# Second plot
ax2.hist(data_b, bins=60, density=False, alpha=0.7, color="b", label=f'Mean=10, Std Dev=50')
ax2.set_title('Normal Distribution [Mean=10, Std Dev=50]')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')
ax2.legend()

plt.tight_layout()
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
def list_number_exponential(mean, size):
    return np.random.exponential(scale=mean, size=size)

# Generate data
data_1 = list_number_exponential(10, 10)
data_2 = list_number_exponential(10, 100)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot histograms
ax1.hist(data_1, bins=100, density=True, alpha=0.7, color="b", label="mean=10, size=10")
ax1.set_xlabel("Value")
ax1.set_ylabel("Density")
ax1.set_title("Exponential Distribution (size=10)")
ax1.legend()

ax2.hist(data_2, bins=100, density=True, alpha=0.7, color="b", label="mean=10, size=100")
ax2.set_xlabel("Value")
ax2.set_ylabel("Density")
ax2.set_title("Exponential Distribution (size=100)")
ax2.legend()

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia.
